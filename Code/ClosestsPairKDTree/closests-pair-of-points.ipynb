```python
from sklearn.neighbors import KDTree
from numpy import random
import numpy as np
import time

# size(number of points, number of dimensions)
points = random.randint(1000, size = (10000, 10))

#leaf size indicates algorithm indicates when algo switches to brute force
tree = KDTree(points, leaf_size = 2)

#here k = 2 means how much points we want
start = time.time()
nearest_dist, nearest_ind = tree.query(points, k=2)
finish = time.time()



i = int(0)
min_dist = float("inf")
for dist in nearest_dist[:, 1]:
    
    if dist < min_dist:
        min_dist = dist
        min_nodes = nearest_ind[i]
    i += 1

print('The minimum distance is:')
print(min_dist)

print(f"Between the points")

for i in range(2):
    print(f"Point Index : {min_nodes[i]} and Co-ordinates of points : {points[min_nodes[i]]}") 

print(f"Total execution time is  :  {finish - start}")
```
